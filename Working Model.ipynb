```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
Shroom_Stats = pd.read_csv('Resources/mushrooms.csv')
```

```python
Shroom_Stats.head()
```

```python
Shroom_Stats.columns
```

```python
labelencoder=LabelEncoder()
for column in Shroom_Stats.columns:
    Shroom_Stats[column] = labelencoder.fit_transform(Shroom_Stats[column])
```

```python
Shroom_Stats
```

```python
CShape = Shroom_Stats["cap-shape"]
CSurface = Shroom_Stats["cap-surface"]
CColor = Shroom_Stats["cap-color"]
Bruises = Shroom_Stats["bruises"]
Odor = Shroom_Stats["odor"]
GAttach = Shroom_Stats["gill-attachment"]
GSpace = Shroom_Stats["gill-spacing"]
GSize = Shroom_Stats["gill-size"]
GColor = Shroom_Stats["gill-color"]
SSBR = Shroom_Stats["stalk-surface-below-ring"]
SCAR = Shroom_Stats["stalk-color-above-ring"]
SCBR = Shroom_Stats["stalk-color-below-ring"]
Veil = Shroom_Stats["veil-color"]
Rnum = Shroom_Stats["ring-number"]
Rtype = Shroom_Stats["ring-type"]
SporePC = Shroom_Stats["spore-print-color"]
Edible = Shroom_Stats["class"].values.reshape(-1,1)
```

```python
X = Shroom_Stats[["cap-shape", "cap-surface", "cap-color", "bruises", "odor", "gill-attachment", "gill-spacing", "gill-size", "gill-color", "stalk-surface-below-ring", "stalk-color-above-ring", "stalk-color-below-ring", "veil-color", "ring-number", "ring-type", "spore-print-color"]]
y = Edible
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```
