```python
from numpy.random import seed
seed(1)
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
Shroom_Stats = pd.read_csv('Resources/mushrooms.csv')
```

```python
labelencoder=LabelEncoder()
for column in Shroom_Stats.columns:
    Shroom_Stats[column] = labelencoder.fit_transform(Shroom_Stats[column])
```

```python
Shroom_Stats
```

```python
CShape = Shroom_Stats["cap-shape"]
CSurface = Shroom_Stats["cap-surface"]
CColor = Shroom_Stats["cap-color"]
Bruises = Shroom_Stats["bruises"]
Odor = Shroom_Stats["odor"]
GAttach = Shroom_Stats["gill-attachment"]
GSpace = Shroom_Stats["gill-spacing"]
GSize = Shroom_Stats["gill-size"]
GColor = Shroom_Stats["gill-color"]
SSBR = Shroom_Stats["stalk-surface-below-ring"]
SCAR = Shroom_Stats["stalk-color-above-ring"]
SCBR = Shroom_Stats["stalk-color-below-ring"]
Veil = Shroom_Stats["veil-color"]
Rnum = Shroom_Stats["ring-number"]
Rtype = Shroom_Stats["ring-type"]
SporePC = Shroom_Stats["spore-print-color"]
Edible = Shroom_Stats["class"]
```

```python
X = Shroom_Stats[["cap-shape", "cap-surface", "cap-color", "bruises", "odor", "gill-attachment", "gill-spacing", "gill-size", "gill-color", "stalk-surface-below-ring", "stalk-color-above-ring", "stalk-color-below-ring", "veil-color", "ring-number", "ring-type", "spore-print-color"]]
y = Edible
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=16))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:20])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:20])}")
```
